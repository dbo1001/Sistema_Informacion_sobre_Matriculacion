### Login o pantalla principal de la aplicación
   
#### Usuario: admin
#### Contraseña: admin

    1. Se muestra una pantalla donde se pide que se introduzca un Usuario y una Contraseña.
    2. Una vez introducido un Usuario y una Contraseña, se pueden dar 4 casos:
        2.1. Si tanto el Usuario como la contraseña son incorrectos, se muestra un mensaje de:
               Acceso denegado  (rojo)
               Usuario y Contraseña incorrectos  (rojo)
        2.2. Si únicamente el Usuario es incorrecto, se muestra un mensaje de:
               Acceso denegado  (rojo)
               Usuario incorrecto  (rojo)
        2.3. Si únicamente la Contraseña es incorrecta, se muestra un mensaje de:
               Acceso denegado  (rojo)
               Contraseña incorrecta  (rojo)
        2.4. Si tanto el Usuario como la contraseña son correctos, se muestra un mensaje:
               Acceso permitido  (verde)
               ¡Bienvenido!  (verde)
    3. Cuando se cumple el caso 2.4. se accede a la aplicación.

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

from tkinter import *
from tkinter import ttk, font
import getpass
#import time

# Gestor de geometría (place)

class Aplicacion():
    def __init__(self):
        self.raiz = Tk()
        
        self.raiz.geometry("850x550") # Se define la dimensión de la ventana
        self.raiz.config(bg = "powder blue") # Asignarle color de fondo
        self.raiz.resizable(0,0)  # Se establece que no se pueda cambiar el tamaño de la ventana
        self.raiz.title("Acceso a Data WareHouse")
        self.raiz.iconbitmap("logo.ico") 
        
        mensajeIntroduce = Label(text="Introduzca Usuario y Contraseña para poder acceder a Data WareHouse", fg="black", font=("Comic Sans MS", 18))
        mensajeIntroduce.place(x=10, y=20) 
        
        self.fuente = font.Font(weight='bold', font=("Comic Sans MS", 12))                       
        self.etiq1 = ttk.Label(self.raiz, text="Usuario:", font=self.fuente)
        self.etiq2 = ttk.Label(self.raiz, text="Contraseña:", font=self.fuente)
                               
        # Se definen 2 variables cadenas que se asigna a la opción 'textvariable' de un widget 'Label' para
        # mostrar mensajes en la pantalla. Se asigna el color negro(por defecto) a la opción 'foreground' para el mensaje. 
        self.mensa = StringVar()
        self.mensa2 = StringVar()
        self.etiq3 = Label(self.raiz, textvariable=self.mensa, font=self.fuente, foreground='black')
        self.etiq4 = Label(self.raiz, textvariable=self.mensa2, font=self.fuente, foreground='black')
                     
        self.usuario = StringVar()
        self.clave = StringVar()
        self.usuario.set(getpass.getuser())
        self.ctext1 = ttk.Entry(self.raiz, textvariable=self.usuario, width=30)
        self.ctext2 = ttk.Entry(self.raiz, textvariable=self.clave, width=30, show="*")
        #self.separ1 = ttk.Separator(self.raiz, orient=HORIZONTAL)
        
        self.botonAceptar = ttk.Button(self.raiz, text="Aceptar", padding=(6,6), command = self.aceptar)
        self.botonCancelar = ttk.Button(self.raiz, text="Cancelar", padding=(6,6), command = quit) # command=quit NO FUNCIONA 
            
        # Se definen las ubicaciones de los widgets en la ventana asignando los valores de las opciones 'x' e 'y' en píxeles.                           
        self.etiq1.place(x=300, y=140) # Usuario
        self.etiq2.place(x=300, y=180) # Contraseña
        self.ctext1.place(x=400, y=143) # Caja para escribir el Usuario
        self.ctext2.place(x=400, y=183) # Caja para escribir la Contraseña
        self.etiq3.place(x=360, y=250) # Mensaje a mostrar en rojo/verde
        self.etiq4.place(x=360, y=290)
        #self.separ1.place(x=5, y=145, bordermode=OUTSIDE, height=10, width=420)
        self.botonAceptar.place(x=310, y=350) # Botón de Aceptar
        self.botonCancelar.place(x=410, y=350) # Botón de Cancelar

        
        self.ctext2.focus_set()
        
        # El método 'bind()' asocia el evento de 'hacer clic 
        # con el botón izquierdo del ratón en la caja de entrada
        # de la contraseña' expresado con '<button-1>' con el 
        # método 'self.borrar_mensa' que borra el mensaje y la 
        # contraseña y devuelve el foco al mismo control.
        # Otros ejemplos de acciones que se pueden capturar:
        # <double-button-1>, <buttonrelease-1>, <enter>, <leave>,
        # <focusin>, <focusout>, <return>, <shift-up>, <key-f10>, 
        # <key-space>, <key-print>, <keypress-h>, etc.
        
        #self.ctext2.bind('<button-1>', self.borrar_mensa)
        self.raiz.mainloop()
        
    # Declara método para validar la contraseña y mostrar un mensaje en la propia ventana, utilizando la etiqueta
    # 'self.mensa'. Cuando la contraseña es correcta se asigna el color verde a la etiqueta 'self.etiq3' y
    # cuando es incorrecta el color rojo. Para ello se emplea el método 'configure()' que permite cambiar los valores
    # de las opciones de los widgets.
    
    def aceptar(self):
        if self.clave.get() == 'admin' and self.usuario.get() == 'admin':
            self.etiq3.configure(foreground='green')
            self.etiq4.configure(foreground='green')
            self.mensa.set("Acceso permitido")
            self.mensa2.set("¡Bienvenido!")
            #time.sleep(1) #Esperar 2 segundos y ...
            # Enviar a la página principal de la aplicación
            
        else:
            self.etiq3.configure(foreground='red')
            self.etiq4.configure(foreground='red')
            self.mensa.set("Acceso denegado")
            if self.clave.get() != 'admin' and self.usuario.get() != 'admin':
                self.mensa2.set("Usuario y Contraseña incorrectos")
            elif self.clave.get() != 'admin':
                self.mensa2.set("Contraseña incorrecta")
            elif self.usuario.get() != 'admin':
                self.mensa2.set("Usuario incorrecto")
                
            #self.clave.set("")  # Borra la contraseña que hemos introducido de manera incorrecta
            #self.usuario.set("") # Borra el usuario que hemos introducido de manera incorrecta
            
    # Se declara un método para borrar el mensaje anterior y la caja de entrada de la contraseña
    def borrar_mensa(self, evento):
        self.clave.set("")
        self.mensa.set("")

def main():
    mi_app = Aplicacion()
    return 0

if __name__ == '__main__':
    main()
```
